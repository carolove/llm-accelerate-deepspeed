```python
# 使用load_dataset 加载数据 ("gsm8k", "main")
from datasets import load_dataset

dataset = load_dataset("gsm8k", "main")
print(dataset)
```

```python
# 使用AutoTokenizer 加载tokenizer
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("/home/chendan/Desktop/workspace/workspace/checkpoints/Llama-3___2-1B")
print(tokenizer)
```

```python
# 定义一个tokenize_function
def tokenize_function(examples):
    return tokenizer(examples["question"], examples["answer"], truncation=True, max_length=None)

# 使用map方法应用tokenize_function到dataset的train和test,并且移除question和answer列
dataset_tokenized = dataset.map(tokenize_function, batched=True, remove_columns=["question", "answer"])
# 打印dataset_tokenized的结构
print(dataset_tokenized)
# 打印dataset_tokenized的train的第一个元素的input_ids和attention_mask
print(dataset_tokenized["train"][0]["input_ids"])
print(dataset_tokenized["train"][0]["attention_mask"])
```
